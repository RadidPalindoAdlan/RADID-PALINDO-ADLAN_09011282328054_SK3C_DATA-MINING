```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("d:\\semester3\\DATA MINING\\heart_disease_uci.csv")

print("5 baris pertama:")
print(df.head())
print("\n5 baris terakhir:")
print(df.tail())

```

```python
print("\nInformasi dataset:")
print(df.info())
```

```python
print("\nNilai missing:")
print(df.isnull().sum())
```

```python
print(df.describe())
```

```python
mean_values = df.mean(numeric_only=True)
print(mean_values)
```

```python
median_values = df.median(numeric_only=True)
print(median_values)
```

```python
mode_values = df.mode(numeric_only=True)
print(mode_values)
```

```python
stdv_values = df.std(numeric_only=True)
print(stdv_values)
```

```python
var_values = df.var(numeric_only=True)
print(var_values)
```

```python
skew_values = df.skew(numeric_only=True)
print(skew_values)
```

```python
kurt_values = df.kurt(numeric_only=True)
print(kurt_values)
```

```python
numeric_columns = df.select_dtypes(include=[np.number]).columns

korelasi_matrix = df[numeric_columns].corr()

plt.figure(figsize=(12, 10))
sns.heatmap(korelasi_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix of Numeric Features')
plt.show()
```

```python
df.hist (bins= 30, figsize=(20, 15))
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=df)
plt.xticks(rotation=90)
plt.show()
```

```python
print(df.columns)
```

```python
target_column ='num'
visualized_feature = ['age', 'trestbps','chol','thalch','oldpeak']

sns.pairplot(df[visualized_feature + [target_column]], hue=target_column)
plt.suptitle('Pair Plot untuk Fitur Utamanya', y=1.02)
plt.show()

fig, axes = plt.subplots(2,2, figsize=(15,15))
axes = axes.ravel()

for i,feature in enumerate(visualized_feature[:4]):
    sns.scatterplot(data=df, x=feature, y='thalch',hue=target_column, ax=axes[i])
    axes[i].set_title(f'(feature) vs thalch')

plt.tight_layout()
plt.show()
```
